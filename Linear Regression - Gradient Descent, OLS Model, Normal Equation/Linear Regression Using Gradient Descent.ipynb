## Steps for Gradient Descent:

##### Step 1: Initialize the $ \Theta $ with random values and calculate error

##### Step 2: Calculate the gradient i.e. change in SSE when the $ \Theta $ are changed by a very small value from their original randomly initialized value. This helps us move the values of $ \Theta $ in the direction in which error is minimized.

##### Step 3: Adjust the weights with the gradients to reach the optimal values where SSE is minimized

##### Step 4: Use the new weights for prediction and to calculate the new error

##### Step 5: Repeat steps 2 and 3 till further adjustments to $ \Theta $ doesnt significantly reduce the Error

#### After iterating through a considerable times we will come to a point where the partial derivative of the Cost Function at that particular point will be equivalent to change even the $ \Theta $ values are changed

```python
import numpy as np
X = np.random.rand(100,1) # Generating Random Set of feature values.
y = 3 + 2 * X  + np.random.randn(100,1) # Function to determine y values for a set of feature values.
X = np.c_[np.ones((100,1)),X]
```

```python
learning_rate = 0.1
n_iterations = 1000
m = len(X)
theta_best = np.random.randn(2,1)
for iteration in range(n_iterations):
    gradients = (1/m) * (X.T.dot(X.dot(theta_best) - y))
    theta_best = theta_best - learning_rate * gradients
```
