```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler 
from sklearn.linear_model  import Ridge,Lasso,RidgeCV, LassoCV, ElasticNet, ElasticNetCV, LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import statsmodels.api as sm
import pickle as pk
import warnings
warnings.filterwarnings('ignore')
```

```python
boston = load_boston()
```

```python
data = pd.DataFrame( data = boston.data, columns= (boston.feature_names)).join(pd.DataFrame(boston.target, columns=['Target']))
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
# Visualizing the linearity of independent varibales with our Target variable

plt.figure(figsize=(20,30), facecolor='white')
plotnumber = 1

for column in data.columns:
    if column != 'Target':
        if plotnumber<=len(data.columns) :
            ax = plt.subplot(5,3,plotnumber)
            plt.scatter(data[column],data.Target)
            plt.xlabel(column,fontsize=20)
            plt.ylabel('Price',fontsize=20)
        plotnumber+=1
plt.tight_layout()
```

```python
X = data.iloc[:,data.columns != 'Target']
y = data.Target
```

```python
X.head()
```

```python
y.head()
```

```python
# Evaluating the relevant features for the Model using OLS Model

model_ols = sm.OLS(y,X)
ols_result = model_ols.fit()
ols_result.summary()
```

#### F Statistic
##### H0 : There is no dependency between the Independent Variables and Dependent Variables
##### H1 : There is dependency between the Independent Variables and Dependent Variables
##### F value is larger and Probability of 5 is 0
##### F value and Probability of F value shows us we have libear relationship between Independent and Dependent Variable

### Selecting Relevant Columns
#### T test and Probability of T Test
##### H0 : Coefficient of feature value will be equal to 0
##### H1 : Coefficient of feature value will not be equal to 0
##### Check for p values of T Test at an signifance of 95% i.e. more less than 0.05. Any such columns rejecting that
##### Higher the t value higher the chances of rejecting the H0

```python
ols_selected_col = ['CRIM','ZN','CHAS','RM','DIS','PTRATIO','B','LSTAT']
X = data[ols_selected_col]
X.head()
```

```python
### Evaluating the Data Distribution for the Selected Columns

# Visualizing the linearity of selected independent varibales with our Target variable

plt.figure(figsize=(20,30), facecolor='white')
plotnumber = 1

for column in data[ols_selected_col]:
    if column != 'Target':
        if plotnumber<=len(data.columns) :
            ax = plt.subplot(5,3,plotnumber)
            plt.scatter(data[column],data.Target)
            plt.xlabel(column,fontsize=20)
            plt.ylabel('Price',fontsize=20)
        plotnumber+=1
plt.tight_layout()
```

### Inference from the above plot

##### Chas is a categorical feature
##### RIM, DIS, LSTAT seems to have Linear Relationship with the Target Variable

```python
plt.figure(figsize=(20,40), facecolor='white')
plotnumber = 1

for column in data[ols_selected_col]:
    if column != 'Target':
        if plotnumber<=len(data.columns) :
            ax = plt.subplot(5,3,plotnumber)
            sns.regplot(data[column],data.Target)
            plt.xlabel(column,fontsize=20)
            plt.ylabel('Price',fontsize=20)
        plotnumber+=1
plt.tight_layout()
```

```python
# Checking for Correlation with the Dependent Variable

for i in ols_selected_col:
    if i != 'Target':
        print('{} correlation with Target is {}'.format(i, data[i].corr(data.Target)))
```

### Multicollinearity Evaluation

```python
# Drawing heatmap

plt.subplots(figsize = (14,10))
sns.heatmap(data[ols_selected_col].corr(),annot=True, annot_kws={'size' : 10})
```

```python
# Using Variance Inflation Factor to determine Multi Collinearity

from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = np.array(X)

# we create a new data frame which will include all the VIFs
# note that each variable has its own variance inflation factor as this measure is variable specific (not model specific)


vif = pd.DataFrame()

# here we make use of the variance_inflation_factor, which will basically output the respective VIFs 

vif["VIF"] = [variance_inflation_factor(variables, i) for i in range(variables.shape[1])]
# Finally, I like to include names so it is easier to explore the result
vif["Features"] = X.columns

vif
```

```python
X = data[ols_selected_col]
```

```python
scale = StandardScaler()
X_scaled = scale.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled,y, test_size = 0.25, random_state = 120)
```

```python
model_reg = LinearRegression()
model_reg.fit(X_train,y_train)
```

```python
print('Train Data Score: {}'.format(model_reg.score(X_train,y_train)))
print('Test Data Score: {}'.format(model_reg.score(X_test,y_test)))
```

```python
y_predict = model_reg.predict(X_test)
```

```python
print('Linear Regression R2 Score: {}'. format(r2_score(y_test,y_predict)))
```

```python
model_reg.coef_
```

### Using Lasso Regression to remove the heavy coefficient features

```python
# Finding the best alpha value for the Model

lasscv = LassoCV(alphas = None,cv = 10, max_iter = 100000, normalize = True)
lasscv.fit(X_train, y_train)
```

```python
lasso_reg = Lasso(lasscv.alpha_)
lasso_reg.fit(X_train, y_train)
```

```python
lasso_reg.score(X_test, y_test)
```

```python
lasso_reg.score(X_train, y_train)
```
