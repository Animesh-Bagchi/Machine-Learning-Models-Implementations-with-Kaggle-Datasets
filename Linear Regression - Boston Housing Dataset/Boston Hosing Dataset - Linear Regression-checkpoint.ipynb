```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler 
from sklearn.linear_model  import Ridge,Lasso,RidgeCV, LassoCV, ElasticNet, ElasticNetCV, LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
import pickle as pk
```

```python
boston = load_boston()
```

```python
data = pd.DataFrame( data = boston.data, columns= (boston.feature_names)).join(pd.DataFrame(boston.target, columns=['Target']))
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
# Selecting columns with strong correlation with the Dependent Variable

strong_corr_columns = []
for i in data.columns:
    if i != 'Target':
        if data[i].corr(data.Target) > (0.4) or data[i].corr(data.Target) < (-0.4):
            #print(data[i].corr(data.Target))
            strong_corr_columns.append(i)
strong_corr_columns
```

```python
# Visualizing the linearity of independent strongly correlated varibales with our Target variable

plt.figure(figsize=(20,30), facecolor='white')
plotnumber = 1

for column in data[strong_corr_columns]:
    if plotnumber<=len(strong_corr_columns) :
        ax = plt.subplot(5,3,plotnumber)
        plt.scatter(data[column],data.Target)
        plt.xlabel(column,fontsize=20)
        plt.ylabel('Price',fontsize=20)
    plotnumber+=1
plt.tight_layout()
```

```python
# Drawing heatmap for checking the Multicollinearity

plt.subplots(figsize = (14,10))
sns.heatmap(data[strong_corr_columns].corr(),annot=True, annot_kws={'size' : 10})
```

```python
for i in strong_corr_columns:
    if i != 'Target':
        print('{} correlation with Target is {}'.format(i, data[i].corr(data.Target)))
```

```python
X = data.loc[:,strong_corr_columns]
y = data.Target
```

```python
X.head()
```

```python
y.head()
```

```python
scale = StandardScaler()
X_scaled = scale.fit_transform(X)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = X_scaled

# we create a new data frame which will include all the VIFs
# note that each variable has its own variance inflation factor as this measure is variable specific (not model specific)
# we do not include categorical values for mulitcollinearity as they do not provide much information as numerical ones do
vif = pd.DataFrame()

# here we make use of the variance_inflation_factor, which will basically output the respective VIFs 
vif["VIF"] = [variance_inflation_factor(variables, i) for i in range(variables.shape[1])]
# Finally, I like to include names so it is easier to explore the result
vif["Features"] = X.columns
```

```python
vif
```

```python
# Viewing the data distribution for selected column

plt.figure(figsize=(20,25), facecolor='white')
plotnumber = 1

for column in data[strong_corr_columns]:
    if plotnumber<=14 :
        ax = plt.subplot(4,4,plotnumber)
        sns.distplot(data[column])
        plt.xlabel(column,fontsize=20)
    plotnumber+=1
plt.tight_layout()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_scaled,y,test_size = 0.25,random_state=120)
model_reg = LinearRegression()
model_reg.fit(X_train,y_train)
y_linear_reg_predict = model_reg.predict(X_test)
```

```python
model_reg.score(X_train,y_train)
```

```python
# Let's create a function to create adjusted R-Squared

def adj_r2(x,y):
    r2 = model_reg.score(x,y)
    n = x.shape[0]
    p = x.shape[1]
    adjusted_r2 = 1-(1-r2)*(n-1)/(n-p-1)
    return adjusted_r2
```

```python
adj_r2(X_train,y_train)
```

```python
model_reg.score(X_test,y_test)
```

```python
lasscv = LassoCV(alphas = None,cv = 10, max_iter = 100000, normalize = True)
lasscv.fit(X_train, y_train)
```

```python
# Using Ridge regression model
# RidgeCV will return best alpha and coefficients after performing 10 cross validations. 
# We will pass an array of random numbers for ridgeCV to select best alpha from them

alphas = np.random.uniform(low=0, high=10, size=(50,))
ridgecv = RidgeCV(alphas = alphas,cv=10,normalize = True)
ridgecv.fit(X_train, y_train)
```

```python
ridgecv.alpha_
```

```python
ridge_model = Ridge(alpha=ridgecv.alpha_)
ridge_model.fit(X_train, y_train)
```

```python
ridge_model.score(X_test, y_test)
```
